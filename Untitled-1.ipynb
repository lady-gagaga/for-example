```python
import pandas as pd
import numpy as np
import scipy.stats as ss
import matplotlib
from matplotlib import pyplot as plt
import sklearn
import openpyxl
import statsmodels.api as sm
import seaborn as sns
import statsmodels.stats.proportion
```

библиотеки

```python
df = pd.read_excel('data_10.xlsx')
df = df.transpose()
df.reset_index(inplace=True)
df.rename(columns={'index': 'value'}, inplace=True)
```

```python
grid = np.linspace(2, 8, 200)  

plt.figure(figsize=(16, 7))
plt.hist(df, bins=30, density=True, 
         alpha=0.6, label='Гистограмма выборки') 
plt.plot(grid, ss.expon.pdf(grid, loc=2.2, scale=0.95), color='green', 
         lw=5, label='Плотность случайной величины') 
plt.legend(fontsize=14, loc=1)
plt.grid(ls=':')
plt.show()
```

Сетка для графика

```python
fig = plt.figure()
ax_norm = fig.add_subplot(221)
res = ss.probplot(df['value'], dist=ss.expon, sparams=(
        np.mean(df['value']), np.var(df['value'])), plot=ax_norm)
ax_norm.set_title(
        "QQ-plot for expon distribution", fontsize=10)

fig.tight_layout()
plt.show()
```

Выбрал наиболее подходящее распределение по методу перебора

```python
ss.kstest(df["value"], cdf=ss.expon(ss.expon.fit(df["value"])[0],ss.expon.fit(df["value"])[1]).cdf)
```

Критерий согласия Колмогорова

90% доверительный интервал

```python
a = ss.norm.interval(alpha=0.99, loc=ss.norm.fit(df)[0], scale=ss.norm.fit(df)[1]) 
b = ss.norm.interval(alpha=0.95, loc=ss.norm.fit(df)[0], scale=ss.norm.fit(df)[1])
c = ss.norm.interval(alpha=0.90, loc=ss.norm.fit(df)[0], scale=ss.norm.fit(df)[1])
```

```python
xxx = (df,)  # samples must be in a sequence
res1 = ss.bootstrap(xxx, np.mean, confidence_level=0.99)
res1 = res1.confidence_interval
res2 = ss.bootstrap(xxx, np.mean, confidence_level=0.95)
res2 = res2.confidence_interval
res3 = ss.bootstrap(xxx, np.mean, confidence_level=0.90)
res3 = res3.confidence_interval
```

```python
fig, axs = plt.subplots(nrows= 3 , ncols= 1 )


axs[0].hist(df, alpha=0.3)
axs[0].plot([res1[0],res1[1]],[0,0],'-o',linewidth= 10)
axs[1].hist(df, alpha=0.3)
axs[1].plot([res2[0],res2[1]],[0,0],'-o',linewidth= 10)
axs[2].hist(df, alpha=0.3)
axs[2].plot([res3[0],res3[1]],[0,0],'-o',linewidth= 10)
```
